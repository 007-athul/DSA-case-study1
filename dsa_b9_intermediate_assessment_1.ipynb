## Challenge 2 : Analyze daily expenses and extract insights.
User enters:

Number of transactions
Category and amount

Food 200
Travel 500
Food 300
Shopping 700

### Output
* Total expense
* Highest expense category
* Category-wise total

```python
def analyze_expenses():
    n = int(input("Enter number of transactions: "))

    expenses = {}

    for _ in range(n):
        category, amount = input().split()
        amount = float(amount)

        if category in expenses:
          expenses[category] += amount

        else :
          expenses[category] =amount

    total= sum(expenses.values())

    highest_amount= max(expenses.values())

    for category in expenses:
      if expenses[category] == highest_amount:
        highest_category=category
        break




    print("Total Expense:", total)
    print("Highest Category:", highest_category)
    print("Category-wise:", expenses)

analyze_expenses()
```

### Challenge 3: Data Cleaner (Mini Pandas Logic Without Pandas)

10 ,20 ,abc ,30 ,-5 , 40

* Remove invalid values
* Remove negative values


Show:
* Clean list
* Average
* Max value

```python
def clean_data():
    n = int(input("Enter number of values: "))

    data = []

    for _ in range(n):
        value = input()

        try:
            num = float(value)

            if num < 0:
              continue

            data.append(num)

        except ValueError :
          pass



    if len(data) > 0:
        avg = sum(data) / len(data)
        max_val = max(data)
    else:
        avg = 0
        max_val = None

    print("Clean Data:", data)
    print("Average:", avg)
    print("Max:", max_val)

clean_data()
```
