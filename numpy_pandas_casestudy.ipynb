```python
#task1
import pandas as pd
import numpy as np
```

```python
#task 2
data=pd.read_csv('IMDB-Movie-Data.csv')
```

```python
#task 3
#shape
data.shape
```

```python
#head
data.head()
```

```python
#column names
data.columns.tolist()
```

```python
#datatype
data.dtypes
```

```python
#missing values
data.isnull().sum()
```

```python
#task 4
#checking for duplicated
data.duplicated().sum()
```

```python
#removing duplocated values
data=data.drop_duplicates()
```

```python
#task 5
#renaming column name
data=data.rename(columns={'Title':'Movie_Title'})
data.head()
```

```python
#task 6
# to get unique genre names
data['Genre'].unique()
```

```python
#task 7
#genre count per movie
data['Genre_Count']= data['Genre'].str.split(",").str.len()
data[['Movie_Title','Genre','Genre_Count']].head()
```

```python
#task 8
#Handle missing values in 'Revenue (Millions)' and 'Metascore
data['Revenue (Millions)']=data['Revenue (Millions)'].fillna(
    data['Revenue (Millions)'].mean())

data['Metascore']= data['Metascore'].fillna(
    data['Metascore'].mean())

data[['Revenue (Millions)', 'Metascore']].isnull().sum()
```

```python
#task9
#Top 5 movies by Rating
top5= data.sort_values(by='Rating',ascending=False)
top5[['Movie_Title','Rating']].head()
```

```python
#task 10
#Movies with highest/lowest revenue

highest_revenue=data['Revenue (Millions)'].idxmax()
lowest_revenue=data['Revenue (Millions)'].idxmin()
print(f"Highest Revenue :{data.loc[highest_revenue,'Movie_Title']}\n Lowest Revenue :{data.loc[lowest_revenue,'Movie_Title']}")
```

```python
#task11
#Avg. rating by Director
data.groupby('Director')['Rating'].mean()
```

```python
#task12
#Movies per year
data['Year'].value_counts()
```

```python
#task 13
#Movies with rating > 8 & revenue > 100M
data.loc[(data['Rating'] > 8) & (data['Revenue (Millions)'] > 100),'Movie_Title']
```

```python
#task14
#Group by Director: avg Rating, Revenue
data.groupby('Director')[['Rating','Revenue (Millions)']].mean()
```

```python
#task 15
#Group by Year: count, avg Rating
data.groupby('Year')['Rating'].agg(['count', 'mean'])
```

```python
#task16
#Create a new column Profitability:
#a) Revenue divided by Duration
data['Profitability'] = data['Revenue (Millions)'] / data['Runtime (Minutes)']
data.head()
```

```python
#b) Find top 10 most profitable movies
top10 = data.sort_values('Profitability', ascending=False)

print(top10[['Movie_Title', 'Profitability']].head(10))
```

```python
#task17
#17) Create a column that categorizes movies by rating:
#a) Rating < 6: Poor
#b) 6 ≤ Rating < 7.5: Average
#c) Rating ≥ 7.5: Excellent
data['Rating_Category'] = data['Rating'].apply(
    lambda x: 'Poor' if x < 6
    else 'Average' if x < 7.5
    else 'Excellent'
)
print(data[['Movie_Title', 'Rating', 'Rating_Category']].head())
```

```python
#task18
#Create a custom function to score movies based on:
#i) Apply row['Rating'] * 10 + row['Revenue (Millions)'] as High score
#Apply a function to dataset to assign the score

def movie_score(row):
    return row['Rating'] * 10 + row['Revenue (Millions)']

data['Score'] = data.apply(movie_score, axis=1)

print(data[['Movie_Title', 'Score']].head())
```
