Task 6
Rename columns wherever necessary so that all datasets have a common column
name for merging.
Hint:
● Example: Country → country

```python
gdp_data.rename(columns={'Country': 'country'},inplace =True)
lit_data.rename(columns={'name':'country'},inplace=True)

print("GDP columns:", gdp_data.columns.tolist())
print("Literacy Rate columns:", lit_data.columns.tolist())
print("Population columns:", pop_data.columns.tolist())
print("Internet Users columns:", json_data.columns.tolist())
```

Task 7
Convert the required columns into appropriate numeric data types.

```python

pop_data['population'] = pop_data['population'].astype('int64')
gdp_data['GDP'] = gdp_data['GDP'].astype('int64')
json_data['internet_users'] = json_data['internet_users'].astype('int64')
lit_data['literacy_rate'] = lit_data['literacy_rate'].astype('float64')

print(pop_data.dtypes)
print(gdp_data.dtypes)
print(json_data.dtypes)
print(lit_data.dtypes)
```

Task 8
Merge all four datasets into one DataFrame.
Hint:
● Use the common column (country)
● Choose an appropriate join method.

```python
merged_data= pd.merge(pop_data,gdp_data, on='country', how='inner')
merged_data= pd.merge(merged_data,json_data, on='country', how='inner')

final_data= pd.merge(merged_data,lit_data, on='country', how='inner')
final_data.head()
```

Task 9
Create the following calculated columns:
1. Internet Penetration (%)
2. GDP Per Capita

Hint:
Round the calculated values to two decimal places.

```python
final_data['Internet Penetration (%)']=((final_data['internet_users']/ final_data['population'])*100).round(2)
final_data['GDP Per Capita ']=(final_data['GDP']/final_data['population'])
final_data.head()
```

Part D – Load Data
Task 10
Create a SQLite database. Store the final merged dataset into a table named
‘country_statistics’
Hint:
● Use the sqlite3 library.
● Load the DataFrame using pandas.


```python
import sqlite3 as sq

conn= sq.connect('world_data.db')
final_data.to_sql('country_statistics', conn, if_exists='replace', index=False)
conn.close()
print("Database Created sucessfully")
```

Part F – Data Analysis Using SQL
Answer the following questions using SQLite queries.

```python
#11 Display the Top 5 countries by GDP.
conn=sq.connect('world_data.db')
query="SELECT country,GDP FROM country_statistics ORDER BY GDP DESC LIMIT 5;"
top5_gdp=pd.read_sql_query(query,conn)
top5_gdp
```

```python
#12 Find the average literacy rate.
query= "SELECT AVG(literacy_rate) FROM country_statistics;"
avg_literacy_rate=pd.read_sql_query(query,conn)
avg_literacy_rate
```

```python
#13 Display all countries having literacy rate above 90%.
query="SELECT country,literacy_rate FROM country_statistics WHERE literacy_rate>90;"
lit_rate_above90=pd.read_sql_query(query,conn)
lit_rate_above90
```

```python
#14Find the Top 5 countries with the highest internet penetration.
query="SELECT country,[INTERNET PENETRATION (%)] from country_statistics ORDER BY [INTERNET PENETRATION (%)] DESC LIMIT 5;"
top5_intpen=pd.read_sql_query(query,conn)
top5_intpen
```

```python
#15 Calculate the total population.
query="SELECT SUM(population) AS total_population FROM country_statistics;"
tot_pop=pd.read_sql_query(query,conn)
tot_pop
```

```python
#16 Find the country with the highest GDP per capita.
query_16 = "SELECT country, [GDP Per Capita ] FROM country_statistics ORDER BY [GDP Per Capita ] DESC LIMIT 1;"
highest_gdp_capita = pd.read_sql_query(query_16, conn)
highest_gdp_capita
```
