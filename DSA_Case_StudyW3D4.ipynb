```python
def avg_temperature(temperature):
    return sum(temperature) / len(temperature)

def avg_humidity(humidity):
    return sum(humidity) / len(humidity)

def weather_condition(avg_temp,avg_hum):
    if avg_temp > 30 and avg_hum < 60:
        return "Hot and Sunny"
    elif avg_temp > 25 and avg_hum >= 60:
        return "Warm and Humid"
    elif avg_temp < 20:
        return "Cold"
    else:
        return "Moderate"

temperature=[23.2,50,9,4,46,35]
humidity=[70,40,60,80.7,45,23]

avg_temp = avg_temperature(temperature)
avg_hum = avg_humidity(humidity)
condition = weather_condition(avg_temp, avg_hum)

print(f"Average Temperature: {avg_temp:.2f}C")
print(f"Average Humidity: {avg_hum:.2f}C")
print(f"Weather Condition: {condition}")
```
