Generate a 5×4 array with random values between 0 and 1 (float).
● Find the row with the maximum sum of values.
● Normalize each row using the formula:

```python
D=np.random.randint(0,2,20).reshape(5,4)
D
```

```python
#maximum row sum
row_sum= D.sum(axis=1)
max_row_val=np.argmax(row_sum)

print(f"Row with maximum Sum index position: {max_row_val}  ")

```

```python
#normalize each row
row_min = D.min(axis=1, keepdims=True)
row_max = D.max(axis=1, keepdims=True)

normalized = (D - row_min) / (row_max - row_min)

print("\nNormalized Array:")
print(normalized)
```

6. Create a 6×6 NumPy array with values from 1 to 36.
● Extract the middle 4×4 submatrix.
● Then extract all even numbers from this submatrix.

```python
arr = np.arange(1, 37).reshape(6, 6)
middle = arr[1:5, 1:5]
even_numbers = middle[middle % 2 == 0]
print("Original Array:")
print(arr)
print("\nMiddle 4x4 Submatrix:")
print(middle)
print("\nEven Numbers:")
print(even_numbers)
```

7. Generate an array of shape (3, 4) with values from 0 to 11.
● Reshape it to (2, 6)
● Then flatten it into a 1D array
● Convert the 1D array back to shape (4, 3)

```python
arr = np.arange(12).reshape(3, 4)
reshaped = arr.reshape(2, 6)
flattened = reshaped.flatten()
final_array = flattened.reshape(4, 3)
print("Original Array:")
print(arr)
print("\nReshaped to (2,6):")
print(reshaped)
print("\nFlattened Array:")
print(flattened)
print("\nReshaped to (4,3):")
print(final_array)
```

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.array([7, 8, 9])
● Join them into a single array
● Reshape the result into a 3×3 matrix
● Extract the diagonal elements

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.array([7, 8, 9])
joined = np.concatenate((a, b, c))
matrix = joined.reshape(3, 3)
diagonal = np.diag(matrix)
print("Joined Array:")
print(joined)
print("\n3x3 Matrix:")
print(matrix)
print("\nDiagonal Elements:")
print(diagonal)
```

10. Create a NumPy array of integers from 1 to 20.
● Extract all values that are divisible by 3 or 5.
● Count how many such values exist.

```python
arr = np.arange(1, 21)
result = arr[(arr % 3 == 0) | (arr % 5 == 0)]
count = len(result)
print("Numbers divisible by 3 or 5:")
print(result)
print("\nCount:")
print(count)
```
